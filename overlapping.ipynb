```python
import math
import matplotlib.pyplot as plt
```

```python
def getR1R2(a1, a2):
    '''円の面積から半径を求める'''
    if a2 > a1:
        a1, a2 = a2, a1
    return math.sqrt(a1 / math.pi), math.sqrt(a2 / math.pi)
```

```python
def getTheta1(r1, r2, theta2):
    '''2つの辺の長さと1つの角度からθ1を求める'''
    sin_theta2 = math.sin(theta2)
    sin_theta1 = r2 / r1 * sin_theta2
    theta1 = math.asin(sin_theta1)
    return theta1
```

```python
def getD(r1, r2, theta1, theta2):
    '''2つの辺の長さと2つの角度からdを求める'''
    #if t2 < math.pi / 2:
    #    d = r1 * math.cos(theta1) + r2 * math.cos(theta2)
    #else:
    #    d = r1 * math.cos(t1) - r2 * math.cos(math.pi - t2)

    d = r1 * math.cos(theta1) + r2 * math.cos(theta2)
    return d
```

```python
def getEstA12(r1, r2, theta1, theta2):
    '''2つの辺の長さと2つの角度からa12の推定値を求める'''
    sin_theta1 = math.sin(theta1)
    cos_theta1 = math.cos(theta1)
    sin_theta2 = math.sin(theta2)
    cos_theta2 = math.cos(theta2)

    s = r1*r1*(theta1-sin_theta1*cos_theta1) + r2*r2*(theta2-sin_theta2*cos_theta2)
    return s
```

```python
def drawCircle2(r1, r2, d):
    '''円を描く'''
    x1 = 0
    y1 = 0
    x2 = d
    y2 = 0

    plt.rcParams["figure.figsize"] = (10, 10)
    figure, axes = plt.subplots()

    plt.xlim(-100, 400)
    plt.ylim(-100, 100)

    circle1 = plt.Circle((x1, y1), r1, fill=False)
    circle2 = plt.Circle((x2, y2), r2, fill=False)

    axes.set_aspect(1)
    axes.add_artist(circle1)
    axes.add_artist(circle2)

    plt.show()
```

```python
def binarySearch(a1, a2, a12, epsilon, iterMax):
    '''theta2で面積を2分探索する'''

    # a1とa2の大きい方をa1とする
    if a1 < a2:
        a1, a2 = a2, a1

    # 面積から半径を求める
    r1, r2 = getR1R2(a1, a2)
    
    # theta2の最小最大値
    theta2Min, theta2Max = 0, math.pi
    
    # theta2の初期値は真ん中
    theta2 = (theta2Min + theta2Max) / 2.0
    # 最初は面積0ではじめる
    estA12 = 0
    
    # 誤差がε以下になるまで続ける
    while abs(estA12 - a12) >= epsilon:
        # theta2からtheta1、d、estA12を求める
        theta1 = getTheta1(r1, r2, theta2)
        d = getD(r1, r2, theta1, theta2)
        estA12 = getEstA12(r1, r2, theta1, theta2)

        # 現状の表示
        print('r1=' + str(round(r1, 3)) + ', r2=' + str(round(r2, 3)) + ', d=' + str(round(d, 3))
              + ', theta1=' + str(round(math.degrees(theta1), 3))
              + ', theta2=' + str(round(math.degrees(theta2), 3))
              + ', estA12=' + str(round(estA12, 3)) +  ', a12=' + str(round(a12, 3)))
        drawCircle2(r1, r2, d)

        # 値の更新
        # theta2が増加すると、estA12は単調増加
        # estA12がa12より小さければ右に、そうでなければ左に解がある
        if estA12 < a12:
            theta2Min = theta2
        else:
            theta2Max = theta2
        theta2 = (theta2Min + theta2Max) / 2.0
    
    # 最終結果の表示
    print('a1=' + str(a1) + ', a2=' + str(a2) + ', a12=' + str(a12))
    print('r1=' + str(round(r1, 4)) + ', r2=' + str(round(r2, 4)) + ', d=' + str(round(d, 4)))
    print('estA12=' + str(round(estA12, 4)) + ', a12=' + str(a12)
          + ', abs(estA12-a12)=' + str(round(abs(estA12-a12), 5)))
```

```python
binarySearch(256, 500, 156, 0.001, 1000)
```

```python
def linearSearch(a1, a2, a12, iterMax):
    '''theta2で面積を線形探索する'''

    # a1とa2の大きい方をa1とする
    if a1 < a2:
        a1, a2 = a2, a1

    # 面積から半径を求める
    r1, r2 = getR1R2(a1, a2)
    
    # theta2の最小最大値
    theta2Min, theta2Max = 0, math.pi
    # theta2の刻み
    dTheta2 = float(theta2Max - theta2Min) / iterMax
    
    # theta2の初期値は0
    theta2 = 0
    # 最初は面積0ではじめる
    estA12 = 0
    
    # 全部線形探索する
    theta1s = []
    theta2s = []
    ds = []
    estA12s = []
    i = 0
    while i <= iterMax:
        # theta2からtheta1、d、estA12を求める
        theta1 = getTheta1(r1, r2, theta2)
        d = getD(r1, r2, theta1, theta2)
        estA12 = getEstA12(r1, r2, theta1, theta2)

        theta1s.append(theta1)
        theta2s.append(theta2)
        ds.append(d)
        estA12s.append(estA12)
        
        # 現状の表示
        print('r1=' + str(round(r1, 3)) + ', r2=' + str(round(r2, 3)) + ', d=' + str(round(d, 3))
              + ', theta1=' + str(round(math.degrees(theta1), 3))
              + ', theta2=' + str(round(math.degrees(theta2), 3))
              + ', estA12=' + str(round(estA12, 3)) +  ', a12=' + str(round(a12, 3)))
        drawCircle2(r1, r2, d)

        # 値の更新
        theta2 += dTheta2
        i += 1
    
    # 履歴の表示
    print('θ1')
    plt.plot(theta1s)
    plt.show()
    print('θ2')
    plt.plot(theta2s)
    plt.show()
    print('d')
    plt.plot(ds)
    plt.show()
    print('estA12')
    plt.plot(estA12s)
    plt.show()
```

```python
linearSearch(256, 500, 156, 10)
```
