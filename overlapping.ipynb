```python
import math
import matplotlib.pyplot as plt
```

```python
def getR1R2(a1, a2):
    '''半径を求める。
       大きい方をr1にする'''
    if a2 > a1:
        a1, a2 = a2, a1
    return math.sqrt(a1 / math.pi), math.sqrt(a2 / math.pi)
```

```python
def getT1(r1, r2, t2):
    sin_t2 = math.sin(t2)
    sin_t1 = r2 / r1 * sin_t2
    t1 = math.asin(sin_t1)
    return t1
```

```python
def getD(r1, r2, t1, t2):
    if t2 < math.pi / 2:
        d = r1 * math.cos(t1) + r2 * math.cos(t2)
    else:
        d = r1 * math.cos(t1) - r2 * math.cos(math.pi - t2)
    return d
```

```python
def getEstA12(r1, r2, t1, t2):
    sin_t1 = math.sin(t1)
    cos_t1 = math.cos(t1)
    sin_t2 = math.sin(t2)
    cos_t2 = math.cos(t2)

    s = r1*r1*(t1-sin_t1*cos_t1) + r2*r2*(t2-sin_t2*cos_t2)
    return s
```

```python
def drawCircle2(r1, r2, d):
    # print('描画')
    x1 = 0
    y1 = 0
    #x2 = r1 + r2 - d
    x2 = d
    y2 = 0

    plt.rcParams["figure.figsize"] = (10, 10)
    figure, axes = plt.subplots()

    plt.xlim(-100, 400)
    plt.ylim(-100, 100)

    circle1 = plt.Circle((x1, y1), r1, fill=False)
    circle2 = plt.Circle((x2, y2), r2, fill=False)

    axes.set_aspect(1)
    axes.add_artist(circle1)
    axes.add_artist(circle2)

    plt.show()
```
